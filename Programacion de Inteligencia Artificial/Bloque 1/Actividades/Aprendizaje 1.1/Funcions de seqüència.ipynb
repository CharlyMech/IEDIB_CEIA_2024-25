##enumerate

Quan s'itera sobre una seqüència sovint volem tenir accés a l'índex de l'element actual. Una solució seria la següent.

```python
# collection no està definit, aquesta cel·la salta error.
i=0
for value in collection:
    # fer qualque cosa amb value
    i += 1
```

Ara bé, com que això és tan freqüent, Python ofereix una funció que ho resol. El fragment anterior es pot implementar així.

```python
for i, value in enumerate(collection):
  #fer qualque cosa amb value
```

Quan s'indexen dades, un patró útil amb enumerate és calcular un diccionari que relaciona els vlaors d'una seqüència amb les seves posicions a la seqüència.

```python
llista = ["zero","u","dos","tres"]
mapping = {} # diccionari buit
for i, v in enumerate(llista):
  mapping[v]=i

mapping
```

```python
mapping["dos"]
```

##sorted

La funció `sorted` retorna una nova llista ordenada a partir dels elements d'una seqüència.

```python
sorted([4,6,2,5,3,1])
```

```python
sorted('Vilafranca de Bonany')
```

##zip

La funció **zip** aparella els elements d'un nombre de llistes, tuples o d'altres seqüències per crear una llista de tuples.

```python
xifres = [1,2,3]
lletres = ["un","dos","tres"]
zipped = zip(xifres,lletres)
list(zipped)
```

zip pot prendre un nombre arbitrari de seqüències, i el nombre d'elements que produeix el determina la seqüència més curta.

```python
val = [False, True] # indica si el nombre és parell

list(zip(xifres,lletres,val)) #el resultat té longitud 3
```

Una utilització molt freqüent de `zip` és iterar simultàniament sobre múltiples seqüències, combinat potser amb `enumerate`.

```python
for i, (a,b) in enumerate(zip(['one','two','three'],lletres)):
  print('{0}: {1}, {2}.'.format(i+1,a,b))
```

Donada una seqüència aparellada, zip també es pot fer servir per desaparellar-la. Podem pensar que es transforma una llista de fileres en una llista de columnes.

```python
pioners = [('Alan', 'Turing'),('Charles','Babbage')]

noms,cognoms=zip(*pioners)

noms
```

```python
cognoms
```

##reversed

`reversed` itera sobre els elements d'una seqüència en ordre invers

```python
list(reversed(range(10))) # compte enrere
```

`reversed` és un generador. Això vol dir que no crea la seqüència inversa fins que es materialitza, per exemple amb `list` o un bucle `for`. Veurem més endavant què és un generador.
