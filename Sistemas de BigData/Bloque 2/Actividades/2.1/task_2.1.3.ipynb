```python
# Import required libraries and load data from file
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Set average height and standar desviation variables (cm)
average = 164.7
std = 7.1

sample_size = 1000 # Modofy this value to test results with different data
np.random.seed(42)  # FIx the random values for data replication reasons -> Comment this line if you wish
sample_heights = np.round(np.random.normal(average, std, sample_size), 1) # Generate the height sample data

# Create the histogram graphic of the data
plt.figure(figsize=(10, 6))
plt.hist(sample_heights, bins=30, edgecolor='black')
plt.xlabel('Height (cm)')
plt.ylabel('Freq')
plt.title(f'Women\'s height distribution ({sample_size} women sample data)') #Distribució de les alçades de 1000 dones
plt.show()
```

```python
# Calculate the average's confidence intervals for 90%, 95% and 99%
confidence_levels = [0.90, 0.95, 0.99]
for confidence in confidence_levels:
    ci = stats.norm.interval(confidence, loc=np.mean(sample_heights), scale=stats.sem(sample_heights))
    print(f'{int(confidence*100)}% Confidence interval: ({ci[0]:.2f}, {ci[1]:.2f})')
```

```python
# Calculate probabilities:
print(f'Less than 160 cm probability: {stats.norm.cdf(160, loc=average, scale=std) * 100:.2f}%')
print(f'Between 160 and 170 cm probability: {(stats.norm.cdf(170, loc=average, scale=std) - stats.norm.cdf(160, loc=average, scale=std)) * 100:.2f}%')
print(f'Between 170 and 180 cm probability: {(stats.norm.cdf(180, loc=average, scale=std) - stats.norm.cdf(170, loc=average, scale=std)) * 100:.2f}%')
print(f'Greater than 180 cm probability: {(1 - stats.norm.cdf(180, loc=average, scale=std)) * 100:.2f}%')
```
