[Matplotlib Documentation](https://matplotlib.org/stable/)

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)

ax1.hist(np.random.randn(100), bins = 20, color = 'k', alpha= 0.3)
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
ax3.plot(np.random.randn(50).cumsum(),'k--')

plt.show()
```

```python
etiquetes = ['Categoria 1', 'Categoria 2', 'Categoria 3', 'Categoria 4']
valors = [50, 80, 20, 30] 

plt.figure(figsize=(8, 6))
plt.bar(etiquetes, valors, color='skyblue')
plt.title('Gr√†fic de barres verticals (Matplotlib)')
plt.xlabel('Categories')
plt.ylabel('Valors')
plt.show()  
```

```python
plt.figure(figsize=(8, 6))
plt.barh(etiquetes, valors, color='skyblue')
plt.title('Gr√†fic de barres hotizontals (Matplotlib)')
plt.xlabel('Valors')
plt.ylabel('Categories')
plt.show() 
```

```python
plt.figure(figsize=(6, 6))
plt.pie(valors, labels=etiquetes, autopct='%1.1f%%', startangle=90,
   colors=['skyblue', 'lightcoral', 'lightyellow', 'lightgreen'])
plt.title('Gr√†fic circular (Matplotlib)')
plt.show()  
```

[Squarify Documentation](https://github.com/laserson/squarify)

```python
import matplotlib.pyplot as plt
import squarify

# Dades d'exemple
etiquetes = ['Categoria 1', 'Categoria 2', 'Categoria 3', 
  'Categoria 4', 'Categoria 5', 'Categoria 6']
valors = [250, 120, 280, 320, 140, 95]
colors = ['#91DCEA', '#64CDCC', '#5FBB68','#F9D23C', '#F9A729', '#FD6F30']

# Treemap
squarify.plot(sizes = valors, label = etiquetes, color = colors )

# Eliminar els eixos:
plt.axis("off")

plt.show()
```

[WordCloud Documentation](https://amueller.github.io/word_cloud/)

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

text = "En un lugar de la Mancha, de cuyo nombre no quiero acordarme, no ha mucho \
tiempo que viv√≠a un hidalgo de los de lanza en astillero, adarga antigua, \
roc√≠n flaco y galgo corredor."

wc = WordCloud(width = 300, height = 300).generate(text)

# Elimina els eixos i mostra les dades com una imatge
plt.axis("off")
plt.imshow(wc, interpolation = "bilinear")

plt.show()  
```
