```python
import seaborn as sns
```

```python
#Vegem la paleta de colors per defecte
paleta = sns.color_palette()
sns.palplot(paleta)
```

```python
#I ara d'una altra paleta, la pastel
paleta = sns.color_palette("pastel")
sns.palplot(paleta)
```

```python
import pandas as pd

# Carregam les dades de flors iris
df = sns.load_dataset("iris")
df.head()
```

```python
#Assignam un estil al gr√†fic (opcional)
sns.set_style("whitegrid")

# Cream un diagrama de barres
sns.barplot(data=df, x='species', y='petal_length')
```

```python
#Una altra manera d'escriure-ho:
sns.barplot(x=df.species, y=df.petal_length)
```

```python
#Canviam els colors i llevam les barres d'error
sns.barplot(data=df, x='species', y='petal_length', palette='pastel', errorbar=None)
```

```python
#Un exemple amb hue
df2 = sns.load_dataset("tips")
sns.barplot(data=df2, x='day', y='tip', hue='sex')
```

```python
#L'exemple anterior sense warnings
sns.barplot(data=df, x='species', y='petal_length', hue='species', palette='pastel', errorbar=None)
```

```python
#Gr√†fic de dispersi√≥
sns.scatterplot(data=df, x='species', y='petal_length', hue='species', palette='bright')
```

```python
#Gr√†fic de caixes
sns.boxplot(data=df, x='species', y='petal_length', hue='species', palette='bright')
```

```python
#Gr√†fic circular amb matplotlib i paleta de colors de seaborn
import matplotlib.pyplot as plt

etiquetes = ['Categoria 1', 'Categoria 2', 'Categoria 3', 'Categoria 4']
valors = [50, 80, 20, 30]
colors = sns.color_palette('pastel')
sns.set_style("dark")
plt.figure(figsize=(6, 6))
plt.pie(valors, labels=etiquetes, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Gr√†fic circular (Matplotlib i Seaborn)')
plt.show()
```
